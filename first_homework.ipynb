```python
for x in range(41):
    result = x**2 + x + 41
    is_prime = True
    if result <= 1:
        is_prime = False
    else:
        for i in range(2, int(result**0.5) + 1):
            if result % i == 0:
                is_prime = False
                break

    if is_prime:
      prime_status = "простой"
    else:
      "не простой"
    print(f"При x={x}, результат {result} - {prime_status}.")
```

1)	Написать процедуру (функцию) вычисления суммы 1 + 1/2 + 1/3 + …+1/n для заданного числа n. Результат представить в виде несократимой дроби p/q (p, q — натуральные).

```python
import math

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def harmonic_series_sum(n):
    numerator = 0
    denominator = 1

    for i in range(1, n + 1):
        new_numerator = numerator * i + denominator
        new_denominator = denominator * i

        common_divisor = gcd(new_numerator, new_denominator)
        numerator = new_numerator // common_divisor
        denominator = new_denominator // common_divisor

    return numerator, denominator

n = int(input("Введите натуральное число n: "))
numerator, denominator = harmonic_series_sum(n)
print(f"Сумма ряда = {numerator}/{denominator}")
```

2)	Составить функцию, проверяющую, являются ли данные три числа взаимно простыми.

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def are_coprime(a, b, c):
    two_gcd = gcd(a, b)
    three_gcd = gcd(two_gcd, c)
    return three_gcd == 1

a = int(input("Введите первое число: "))
b = int(input("Введите второе число: "))
c = int(input("Введите третье число: "))

if are_coprime(a, b, c):
    print("Данные числа являются взаимно простыми.")
else:
    print("Данные числа не являются взаимно простыми.")
```
